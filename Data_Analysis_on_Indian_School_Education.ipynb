```
# Loading the libraries required for our analysis
import pandas as pd
import numpy as np
```

```
# Uploading the data file into colab
# And giving the dataset a name as "dropout"
dropout= pd.read_csv("/content/dropout-ratio-2012-2015.csv")
```

```
# Printing the first 5 rows in the dataset
dropout.head()
```

```
# Printing the column names in the given dataset
dropout.columns
```

```
# Knowing if there are any null values in the dataset
dropout.isna()
```

```
# Searches if there are any null values in each of the columns
dropout.isna().any()
```

```
# Printing the values for a single state from the dataset
data_tl=dropout[dropout['State_UT']=='Telangana']
print(data_tl.head())
```

```
# Printing the names of the States and Union Territories present along with the index of their row no.
dropout.groupby('State_UT').groups
```

```
# Printing the values in the column for "Primary_Boys"
dropout['Primary_Boys'].values
```

```
# Replacing the null values with '0'
dropout=dropout.replace('NR', 0)
```

```
# Converting the values in the Primary_Boys column to float which is necessary for ensuring consistent data type
dropout['Primary_Boys']=dropout['Primary_Boys'].apply(lambda x: float(x))
```

```
#primary_boys=dropout['Primary_Boys'].values.astype(np.float32)
dropout['Primary_Girls']=dropout['Primary_Girls'].values.astype(np.float32)
```

```
# Printing the values for the "Upper Primary_Boys"
dropout['Upper Primary_Boys'].values
```

```
# Replacing the value 'Uppe_r_Primary' in the first row with '0'
dropout=dropout.replace('Uppe_r_Primary',0)
dropout['Upper Primary_Boys'].values
```

```
# Converting the values in the selected columns to 32-bit floating point numbers for memory efficiency and data consistency

dropout['Upper Primary_Boys']=dropout['Upper Primary_Boys'].values.astype(np.float32)
dropout['Upper Primary_Girls']=dropout['Upper Primary_Girls'].values.astype(np.float32)

```

```
dropout['Secondary _Boys']=dropout['Secondary _Boys'].values.astype(np.float32)
dropout['Secondary _Girls']=dropout['Secondary _Girls'].values.astype(np.float32)

dropout['HrSecondary_Boys']=dropout['HrSecondary_Boys'].values.astype(np.float32)
dropout['HrSecondary_Girls']=dropout['HrSecondary_Girls'].values.astype(np.float32)
```

```
# Loading libraries for Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Describing the data from the dataset Statistically
dropout.describe()
```

```
dropout.info()
```

```
years=data_tl['year'].values;
print(years)
```

```
# PLotting the count for the State/UT column
plt.figure(figsize= (35,15))

sns.countplot(dropout['State_UT'])
plt.title('Count of Entries by State/UT', fontsize=24)
plt.xlabel('State/UT', fontsize=18)
plt.ylabel('Count', fontsize=18)
```

```
# Printing the first 5 rows of data for the state Telangana
data_tl.head()
```

```
# Creating Bar graphs for the data "Primary_Girls"
plt.bar(data_tl['Primary_Girls'],data_tl['year'] )
plt.xlabel('Primary_Girls')
plt.ylabel('Year')
plt.title('Year basis Primary_Girls ratio in Telangana')
plt.show()
```

```
# Plot for the number of 'Secondary_Girls' over the years
plt.bar(dropout['Secondary _Girls'],dropout['year'] )
plt.xlabel('Secondary _Girls')
plt.ylabel('Year')
plt.title('Year basis Secondary _Girls ratio')
plt.show()
```

```
plt.bar(dropout['Primary_Boys'],dropout['year'] )
plt.xlabel('Primary Boys')
plt.ylabel('Year')
plt.title('Year basis primary boys ratio')
plt.show()
```

```
# Plot for the variation of number of 'Upper Primary Girls' in Telangana over the years
plt.bar(data_tl['Upper Primary_Girls'],data_tl['year'] )
plt.xlabel('Upper Primary_Girls')
plt.ylabel('Year')
plt.title('Upper Primary_Girls ratio vs Year ')
plt.show()
```

```
# Plotting the data of 'HrSecondary Girls' distribution
plt.hist(dropout['HrSecondary_Girls'])
```

```
# Outlier detection using boxplots
plt.figure(figsize= (20,15))
plt.subplot(4,4,1)
sns.boxplot(dropout['HrSecondary_Girls'])

plt.subplot(4,4,2)
sns.boxplot(dropout['HrSecondary_Boys'])

plt.subplot(4,4,3)
sns.boxplot(dropout['Upper Primary_Boys'])

plt.subplot(4,4,4)
sns.boxplot(dropout['Upper Primary_Girls'])

plt.subplot(4,4,5)
sns.boxplot(dropout['Secondary _Girls'])

plt.subplot(4,4,6)
sns.boxplot(dropout['Secondary _Boys'])



```

```
# Clipping the values within the range defined by the 5th and 95th percentile to reduce outliers
dropout['HrSecondary_Girls']=dropout['HrSecondary_Girls'].clip(lower=dropout['HrSecondary_Girls'].quantile(0.05), upper=dropout['HrSecondary_Girls'].quantile(0.95))
```

```
dropout['HrSecondary_Boys']=dropout['HrSecondary_Boys'].clip(lower=dropout['HrSecondary_Boys'].quantile(0.05), upper=dropout['HrSecondary_Boys'].quantile(0.95))
```

```
dropout['Upper Primary_Boys']=dropout['Upper Primary_Boys'].clip(lower=dropout['Upper Primary_Boys'].quantile(0.05), upper=dropout['Upper Primary_Boys'].quantile(0.95))
```

```
dropout['Upper Primary_Girls']=dropout['Upper Primary_Girls'].clip(lower=dropout['Upper Primary_Girls'].quantile(0.05), upper=dropout['Upper Primary_Girls'].quantile(0.95))
```

```
dropout['Secondary _Girls']=dropout['Secondary _Girls'].clip(lower=dropout['Secondary _Girls'].quantile(0.05), upper=dropout['Secondary _Girls'].quantile(0.95))
```

```
dropout['Secondary _Boys']=dropout['Secondary _Boys'].clip(lower=dropout['Secondary _Boys'].quantile(0.05), upper=dropout['Secondary _Boys'].quantile(0.95))
```

```
# Outlier detection using boxplots
plt.figure(figsize= (20,15))
plt.subplot(4,4,1)
sns.boxplot(dropout['HrSecondary_Girls'])

plt.subplot(4,4,2)
sns.boxplot(dropout['HrSecondary_Boys'])

plt.subplot(4,4,3)
sns.boxplot(dropout['Upper Primary_Boys'])

plt.subplot(4,4,4)
sns.boxplot(dropout['Upper Primary_Girls'])

plt.subplot(4,4,5)
sns.boxplot(dropout['Secondary _Girls'])

plt.subplot(4,4,6)
sns.boxplot(dropout['Secondary _Boys'])



```

```
# Creating scatter plots for each pair of variables in the Dataframe
sns.pairplot(dropout,diag_kind='kde')
```
